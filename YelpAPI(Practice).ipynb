```python
import json
with open('/Users/siblose/.secret/yelp_api.json') as f:
    login = json.load(f)
login.keys()
```

```python
# import the YelpAPI Class
from yelpapi import YelpAPI
import pandas as pd
# Create an instance with your key
yelp_api = YelpAPI(login['api-key'], timeout_s=5.0)
yelp_api
```

```python
help(yelp_api.search_query)
```

```python
# use our yelp_api variable's search_query method to perform our API call
search_results = yelp_api.search_query(location='Baltimore, MD',
                                       term='crab cakes')
print(type(search_results))
search_results.keys()
```

```python
search_results['total']
```

```python

CrabCakesdf = pd.DataFrame(search_results['businesses'])
CrabCakesdf.head(2) 
```

```python
## how many businesses in our results
len(search_results['businesses'])
```

```python
#Obtain ONLY the first two pages of results
#The Yelp API will only return a "page" of 20 results at a time.
#If we want to get the next page of results, 
#we will perform another API call, but we will add an additional argument called "offset."
```

```python
# use our yelp_api variable's search_query method to perform our API call
search_results = yelp_api.search_query(location='Baltimore, MD',
                                       term='crab cakes',
                                       offset = 20)
```

```python
CrabCakesdf2 = pd.DataFrame(search_results['businesses'])
CrabCakesdf2.head(2)
```

```python
#Concatenate the results into one data frame

## concatenate the previous results and new results. 
businesses = pd.concat([CrabCakesdf, CrabCakesdf2],
                      ignore_index=True)
display(businesses.head(3), businesses.tail(3))
```

```python
##Save the businesses as a records-oriented JSON file. (df.to_json(orient='records'))
businesses.to_json(orient='records')
```
